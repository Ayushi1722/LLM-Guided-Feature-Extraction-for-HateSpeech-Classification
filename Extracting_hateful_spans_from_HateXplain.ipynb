```python
import pandas as pd
import csv
import ast
```

```python
file_path = 'Rationales_file_HateEXPLAIN_VALIDATION_dataset.csv'
dataset = pd.read_csv(file_path)
dataset
```

```python
len(dataset['rationales'])
```

```python
updated_file_path = 'HateExplain_VALIDATION_GPT_HUMAN_Rationales.csv'
dataset = dataset.drop('List_Of_Rationales', axis=1)

with open(updated_file_path, mode='w', newline='') as updated_file:
    csv_writer = csv.writer(updated_file)

    csv_writer.writerow(dataset.columns.tolist() + ['List_Of_Human_Rationales'])

    for index, row in dataset.iterrows():
        post_tokens = row['post_tokens']
        post_tokens = ast.literal_eval(post_tokens)
        rationales = row['rationales']
        rationales = ast.literal_eval(rationales)

        output = []
        for index_row in rationales:
            selected_tokens = [post_tokens[min(i, len(post_tokens)-1)] for i in range(len(index_row)) if index_row[i] == 1]
            output.append(selected_tokens)

        row_values = row.tolist()
        row_values.append(output)

        csv_writer.writerow(row_values)

updated_data = pd.read_csv(updated_file_path)
```
