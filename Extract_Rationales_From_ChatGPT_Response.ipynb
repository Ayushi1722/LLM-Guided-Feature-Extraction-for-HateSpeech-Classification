```python
import json
import csv
import pandas as pd
import time as t

# Create a new CSV file to write the updated data
updated_file_path = 'Rationales_file_GAB_dataset.csv'
with open(updated_file_path, mode='w', newline='') as updated_file:
    csv_writer = csv.writer(updated_file)

    csv_writer.writerow(df.columns.tolist() + ['ChatGPT_Rationales'])

    for index, row in df.iterrows():
        text = row['ChatGPT Response']

        try:
            data = json.loads(text)
        except json.JSONDecodeError:
            # Handle the case where 'text' is not a valid JSON
            print("Json Decoded error for data: ", data ,"\n")
            print("Type of Json Decoded error data: ", type(data), "\n")
            data = {"rationales": [], "derogatory_language": [], "cuss_words": []}

        rationales = "NA" if not data["rationales"] else ", ".join(data["rationales"])
        derogatory_language = "NA" if not data["derogatory_language"] else ", ".join(data["derogatory_language"])
        cuss_words = "NA" if not data["cuss_words"] else ", ".join(data["cuss_words"])

        # Format the string
        formatted_string = f"Rationales: {rationales}; Derogatory Language: {derogatory_language}; Cuss Words: {cuss_words}"
        # print(f"Formatted String: {formatted_string}", "\n")
        row_values = row.tolist()
        row_values.append(formatted_string)
        csv_writer.writerow(row_values)
        t.sleep(5)

# Read the updated CSV file into a new DataFrame
updated_data = pd.read_csv(updated_file_path)
```
